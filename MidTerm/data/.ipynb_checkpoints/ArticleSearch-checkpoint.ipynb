```python
import os
nyt_article_search_key = os.get_env('nyt_article_search_key')
print(nyt_article_search_key)
```

```python
import requests
from pprint import pprint
import json
import time
for i in range(1, 200):
    j=str(i)
    r = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q=obama&page="+j+"&api-key="+nyt_article_search_key)   
    data = r.json()
    time.sleep(5)
    print(len(data))
    pprint(data)
    with open("data/obama"+j+".json", 'w') as outfile:
        json.dump(data, outfile)
```
